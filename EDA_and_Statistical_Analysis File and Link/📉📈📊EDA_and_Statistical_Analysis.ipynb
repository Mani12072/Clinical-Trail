**Getting BMI category and Age_category**

```python
# Classify BMI into categories
def classify_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal weight'
    else:
        return 'Overweight'

# Apply the classification to a new column
patients['bmi_category'] = patients['bmi'].apply(classify_bmi)


import pandas as pd
from datetime import datetime

def categorize_age(birthdate):
    # Convert birthdate to datetime object
    birthdate = pd.to_datetime(birthdate)
    today = datetime.today()
    birthdate = today.year - birthdate.year - ((today.month, today.day) < (birthdate.month, birthdate.day))

    # Categorizing without printing age
    if birthdate <= 12:
        return "Child"
    elif 13 <= birthdate <= 19:
        return "Teen"
    elif 20 <= birthdate <= 64:
        return "Adult"
    else:
        return "Senior"

# Apply the age categorization to a new column
patients['age_category'] = patients['birthdate'].apply(categorize_age)
patients
```


####**Skewness significantly reduced, approaching a normal distribution.**
1.Use Shapiro-Wilk test to Know the it is following normal distribution or not

```python
# Checking for Normality through Statistics

# Shapiro Wilk Test : performed to check wether distribution is normal or not.
from scipy.stats import shapiro


# Perform Shapiro-Wilk test on Height distribution :
patients_height, p_value_height = shapiro(patients['height'])

print("patients Statistic of height : ", patients_height)
print("p-value_height:", p_value_height)

if p_value_height < 0.05:
    print("Reject null hypothesis: Height does not follow a normal distribution")
else:
    print("Fail to reject null hypothesis: height follows a normal distribution")

print()
# Perform Shapiro-Wilk test on Weight distribution :
patients_weight, p_value_weight = shapiro(patients['weight'])

print("patients Statistic of weight : ", patients_weight)
print("p-value_weight:", p_value_weight)

if p_value_weight < 0.05:
    print("Reject null hypothesis: Weight does not follow a normal distribution")
else:
    print("Fail to reject null hypothesis: Weight follows a normal distribution")

print()
# Perform Shapiro-Wilk test on BMI distribution :
patients_bmi, p_value_bmi = shapiro(patients['bmi'])

print("patients Statistic of bmi : ", patients_bmi)
print("p-value_bmi:", p_value_bmi)

if p_value_bmi < 0.05:
    print("Reject null hypothesis: BMI does not follow a normal distribution")
else:
    print("Fail to reject null hypothesis: BMI follows a normal distribution")
```
